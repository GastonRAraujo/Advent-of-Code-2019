```
class amp():

  def __init__(self, posicion, dataSet, phase):
    self.posicion = posicion
    self.dataSet = dataSet
    self.phase = phase
    self.inputCount=0
    
  def cambioCode(self, nuevoCode):
    self.dataSet = nuevoCode  

  def getInfo(self):
    returnVal= self.dataSet[self.posicion]
    return returnVal

  def cambioPos(self, nuevaPos):
    self.posicion = nuevaPos  

  def getPhase(self):
    return self.phase

  def getData(self):
    return self.dataSet  
  
  def getPos(self):
    return self.posicion
```
