```
#Parte 1 la imagen mide 25*6
#--------------------------------DATOS------------------------------------------
with open('/content/drive/My Drive/Colab Notebooks/Day8/puzzleData.txt', 'r') as f: 
    
    data=f.read()

layers=[]


dataLen=len(data)
print("La cantidad de bits es:",dataLen)

n=0
while n < dataLen:
 
  layer=''
  
  for i in range(25*6):
    layer=layer+str(data[n+i])  
    
  layers.append(layer)
  n += 25*6
```

```
def cantCeros(monoLayer):
    cant=0
    for j in range(len(monoLayer)):
      if monoLayer[j] == '0':
        cant += 1
    return cant

def contadorZeros(ArrayLayer):
  
  lowSol=1000000000000000000000000000000
  returnValue = -1

  for k in range(100):
    sol=cantCeros(ArrayLayer[k])
    
    if sol < lowSol:
      
      returnValue = k
      lowSol = sol
  
  return returnValue

def cantNumber(monoLayer, Number):
  cant=0
  for j in range(len(monoLayer)):
       if monoLayer[j] == Number:
        cant += 1
  return cant
```

```
SubIndice = contadorZeros(layers)
print(layers[SubIndice])
print(cantNumber(layers[SubIndice],'0'))
print(cantNumber(layers[SubIndice],'1'))
print(cantNumber(layers[SubIndice],'2'))
print("La solucion es: ", cantNumber(layers[SubIndice],'1')*cantNumber(layers[SubIndice],'2'), "\n y la capa es", SubIndice)
```

```
#Parte 2 decoding the image

# 2 es transparente
# 1 es blanco
# 0 es negro

#--------------------------------DATOS------------------------------------------
with open('/content/drive/My Drive/Colab Notebooks/Day8/puzzleData.txt', 'r') as f: 
    
    data=f.read()

layers=[]


dataLen=len(data)
print("La cantidad de bits es:",dataLen)

n=0
while n < dataLen:
 
  layer=''
  
  for i in range(25*6):
    layer=layer+str(data[n+i])  
    
  layers.append(layer)
  n += 25*6
```

```
def Imaginator3000(ArrayLayer):
  
    mono = ArrayLayer[0]
    returnValue = ''

    for i in range(len(mono)):
      
      j = 1
      aux = mono[i]
      while aux == '2':
        aux = ArrayLayer[j][i]
        j += 1

    
      returnValue += aux

    return returnValue
```

```
imagen=Imaginator3000(layers)

inicio=0
fin=25

for m in range(150):
  print(imagen[inicio:fin])
  inicio += 25
  fin += 25
```

```
array2 = []
for fila in range(6):
  array1 = []
  for columna in range(25):
    array1.append(int(imagen[25*fila+columna])*(columna+1)*-1.02)
  array2.append(array1)

print(array2)
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(10,30))
ax = fig.add_subplot(1,1,1)
ax.imshow(array2, cmap='terrain')
```
